```python
# Q1. Generate all combinations of letters from dictionary values

data = {'1': ['a', 'b'], '2': ['c', 'd']}

keys = list(data.keys())
list1 = data[keys[0]]
list2 = data[keys[1]]

for i in list1:
    for j in list2:
        print(i + j)
```

```python
# Q2. Print top three items by value

shop = {'item1': 45.50, 'item2': 35, 'item3': 41.30,
        'item4': 55, 'item5': 24}

# Convert dictionary to list of (key,value) tuples
items = list(shop.items())

# Manual sorting without using sorted()
for i in range(len(items)):
    for j in range(i + 1, len(items)):
        if items[i][1] < items[j][1]:
            items[i], items[j] = items[j], items[i]

# Print top 3
for k in range(3):
    print(items[k][0], items[k][1])
```

```python
# Q3. Find N maximum values from dictionary

data = {'a': 5, 'b': 14, 'c': 32, 'd': 35, 'e': 24,
        'f': 100, 'g': 57, 'h': 8, 'i': 100}

items = list(data.items())

# Manual sorting (descending)
for i in range(len(items)):
    for j in range(i + 1, len(items)):
        if items[i][1] < items[j][1]:
            items[i], items[j] = items[j], items[i]

# Function to get top N keys
def top_n(n):
    result = []
    for i in range(n):
        result.append(items[i][0])
    return result

print("1 maximum value:", top_n(1))
print("2 maximum values:", top_n(2))
print("5 maximum values:", top_n(5))
```

```python
# Q4. Combine two lists into a dictionary

keys = ['a', 'b', 'c', 'd', 'e', 'f']
values = [1, 2, 3, 4, 5]

result = {}

# Only map until smaller list size
size = min(len(keys), len(values))

for i in range(size):
    result[keys[i]] = values[i]

print(result)
```

```python
# Q5. Mathematical expression evaluator (supports nested dictionaries)

def evaluate(expr, variables={}):
    # If expr is a number → return it
    if type(expr) in (int, float):
        return expr
    
    # If expr is a variable → return variable value
    if type(expr) == str:
        return variables[expr]

    # Otherwise dictionary expression
    left_val = evaluate(expr["left"], variables)
    right_val = evaluate(expr["right"], variables)
    op = expr["op"]

    if op == "+":
        return left_val + right_val
    elif op == "-":
        return left_val - right_val
    elif op == "*":
        return left_val * right_val
    elif op == "/":
        return left_val / right_val


# Example 1
expr = {
    "op": "+",
    "left": {"op": "*", "left": 2, "right": 3},
    "right": 5
}

print("Result:", evaluate(expr))
```
