#### 1.	Write a Python program to create and display all combinations of letters, selecting each letter from a different key in a dictionary.
#### Sample data: {'1':['a','b'], '2':['c','d']}
#### Expected Output:ac ad bc bd

```python
data = {'1': ['a', 'b'], '2': ['c', 'd']}

list1 = data['1']
list2 = data['2']

for x in list1:
    for y in list2:
        print(x + y, end= " ")
```

#### 2.	Write a Python program to get the top three items in a shop.
#### Sample data: {'item1': 45.50, 'item2':35, 'item3': 41.30, 'item4':55, 'item5': 24}
#### Expected Output:
#### item4 55
#### item1 45.5
#### item3 41.3

```python
data = {'item1': 45.50, 'item2':35, 'item3': 41.30, 'item4':55, 'item5': 24}

items = list(data.items())

for i in range(len(items)):
    for j in range(i + 1, len(items)):
        if items[j][1] > items[i][1]:  
            items[i], items[j] = items[j], items[i]

for i in range(3):
    print(items[i][0], items[i][1])
```

#### 4.	Write a Python program to combine two lists into a dictionary. The elements of the first one serve as keys and the elements of the second one serve as values. Each item in the first list must be unique and hashable.
#### Original lists:
#### ['a', 'b', 'c', 'd', 'e', 'f']
#### [1, 2, 3, 4, 5]
#### Combine the values of the said two lists into a dictionary:
#### {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5}

```python
keys = ['a', 'b', 'c', 'd', 'e', 'f']
values = [1, 2, 3, 4, 5]

result = {}

for i in range(len(values)): 
    result[keys[i]] = values[i]

print(result)
```

#### 5.	Write a Python function that evaluates mathematical expressions represented as dictionaries. The expression dictionary should support basic operations (+, -, *, /), variables, and nested expressions.

```python
def evaluate(expr, variables):
    if isinstance(expr, (int, float)):
        return expr

    if isinstance(expr, str):
        return variables[expr]

    op = expr["op"]
    left_value = evaluate(expr["left"], variables)
    right_value = evaluate(expr["right"], variables)

    if op == "+":
        return left_value + right_value
    elif op == "-":
        return left_value - right_value
    elif op == "*":
        return left_value * right_value
    elif op == "/":
        return left_value / right_value
expr = {
    "op": "*",
    "left": "a",
    "right": {
        "op": "+",
        "left": "b",
        "right": 5
    }
}
variables = {"a": 20, "b": 3}
print(evaluate(expr, variables))
```
