#### Q1. Write a program that has classes such as Student, Course and Department. Enroll a student in a course of a particular department.

```python
class Department:
    def __init__(self, name):
        self.name = name

class Course:
    def __init__(self, name, department):
        self.name = name
        self.department = department

class Student:
    def __init__(self, name):
        self.name = name

    def enroll(self, course):
        print(f"{self.name} enrolled in {course.name} ({course.department.name} Department)")

dept = Department("Computer Science")

course = Course("Python Programming", dept)

student = Student("Rahul")

student.enroll(course)
```

#### Q2. Write a program with class Bill. The users have the option to pay the bill either by cheque or by cash. Use the inheritance to model this situation.

```python
class Bill:
    def pay(self):
        print("Bill payment process starts.")

class CashPayment(Bill):
    def pay(self):
        print("Bill paid using Cash.")

class ChequePayment(Bill):
    def pay(self):
        print("Bill paid using Cheque.")
        
print("1. Pay by Cash")
print("2. Pay by Cheque")

choice = int(input("Enter your choice: "))
if choice == 1:
    payment = CashPayment()
elif choice == 2:
    payment = ChequePayment()
else:
    print("Invalid choice!")
    exit()
payment.pay()
```

#### Q3.Write a program to overload the - =operator to subtract two Distance Objects.

```python
class Distance:
    def __init__(self, meters):
        self.meters = meters

    def __isub__(self, other):
        self.meters = self.meters - other.meters
        return self

d1 = Distance(50)
d2 = Distance(20)
print("Before subtraction:", d1.meters)
d1 -= d2  
print("After subtraction:", d1.meters)
```
